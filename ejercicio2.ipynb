```python
from matplotlib import pyplot as plt 
import numpy as np 
  
  
names = ['Marta', 'Carmen', 'Lydia', 
        'Samira', 'Luna', 'Fatima'] 
  
data = [30, 15, 55, 29, 12, 41] 
  
fig = plt.figure(figsize =(15,12)) 
plt.pie(data, labels = names) 
  
plt.show() 
```

```python
import matplotlib.pyplot as plt

x = ['Marta', 'Carmen', 'Lydia', 
        'Samira', 'Luna', 'Fatima']
 
## Declaramos valores para el eje y
y = [50,20,35,47,23,40]
 
## Creamos Gráfica
plt.bar(x, y)
 
plt.ylabel('Cantidad de seguidores')
 
## Legenda en el eje x
plt.xlabel('Influencers')
 
## Título de Gráfica
plt.title('Cantidad de seguidores de las influencers de moda')
 
## Mostramos Gráfica
plt.show()
```
